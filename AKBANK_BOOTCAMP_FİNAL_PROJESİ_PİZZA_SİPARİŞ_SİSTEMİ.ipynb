```python
import csv
from datetime import datetime
import datetime
import os 

if os.path.exists("Orders_Database.csv"):
  os.remove("Orders_Database.csv")

title=['AD', 'PIZZA', 'KREDIKARTINUMARASI', 'PAROLA', 'UCRET', 'ZAMAN']

with open('Orders_database.csv', mode='a', newline='') as file:
  writer = csv.writer(file)
  writer.writerow(title)

pizza_fiyatlari = {
    "Klasik Pizza": 20.0,
    "Margarita Pizza": 35.0,
    "Turk Pizza": 75.0,
    "Dominos Pizza": 30.0,
}
malzeme_fiyatları = {
    "Zeytin": 2.0,
    "Et": 5.5,
    "Kasar peyniri": 8.0,
    "Salam": 8.0,
    "Sogan": 4.0,
    "Mısır": 5.0,
}


if os.path.exists("Menu.txt"):
  os.remove("Menu.txt")    
with open("Menu.txt", "w", encoding="utf-8") as f:
    f.write("PİZZA SİPARİŞ SİSTEMİ \n")
    f.write("##################################\n")
    f.write(" #############\n")
    f.write("|                              |\n")
    f.write("| * Pizza seçin: \n")
    for i, pizza in enumerate(pizza_fiyatlari.keys()):
        f.write(f"| {i+1}: {pizza} ({pizza_fiyatlari[pizza]}₺){' '*(22-len(pizza)-len(str(pizza_fiyatlari[pizza])))}|\n")
    f.write("-#################################-\n")
    f.write("|                              -\n")
    f.write("| * malzeme seçin:   \n")
    for i, malzeme in enumerate(malzeme_fiyatları.keys()):
        f.write(f"| {i+100}: {malzeme} ({malzeme_fiyatları[malzeme]}₺){' '*(22-len(malzeme)-len(str(malzeme_fiyatları[malzeme])))}|\n")
    f.write("|                              |\n")
    f.write("| * Bizi tercih ettiğiniz için teşekkürler |\n")
    f.write("|#################################|\n")
    f.write("bitirmek için 0'a basın")
```

```python
class Pizza:
   
    def __init__(self):
        self.description = "Tanimlanmamiş Pizza"
        self.cost = 0.0
    
    def get_description(self):
        return self.description

     
    def get_cost(self):
        return self.cost


class KlasikPizza(Pizza):
    def __init__(self):
        self.description = "Klasik Pizza"
        self.cost = pizza_fiyatlari["Klasik Pizza"]

class MargaritaPizza(Pizza):
    def __init__(self):
        self.description = "Margarita Pizza"
        self.cost = pizza_fiyatlari["Margarita Pizza"]

class TurkPizza(Pizza):
    def __init__(self):
        self.description = "Turk Pizza"
        self.cost = pizza_fiyatlari["Turk Pizza"]

class DominosPizza(Pizza):
    def __init__(self):
        self.description = "Dominos Pizza"
        self.cost = pizza_fiyatlari["Dominos Pizza"]


class Decorator(Pizza):
    def __init__(self, component):
        self.component = component
    
    
    def get_cost(self):
        return self.component.get_cost() + Pizza.get_cost(self)

   
    def get_description(self):
        return self.component.get_description() + ' ' + Pizza.get_description(self)


class Zeytin(Decorator):
    def __init__(self, component):
        self.component = component
        self.description = "Zeytin"
        self.cost = malzeme_fiyatları["Zeytin"]

class Et(Decorator):
    def __init__(self, component):
        self.component = component
        self.description = "Et"
        self.cost = malzeme_fiyatları["Et"]


class Mantar(Decorator):
    def __init__(self, component):
        self.component = component
        self.description = "Mantar"
        self.cost = malzeme_fiyatları["Mantar"]

class Kasarpeyniri(Decorator):
    def __init__(self, component):
        self.component = component
        self.description = "Kasarpeyniri"
        self.cost = malzeme_fiyatları["Kasarpeyniri"]

class Salam(Decorator):
    def __init__(self, component):
        self.component = component
        self.description = "Salam"
        self.cost = malzeme_fiyatları["Salam"]

class Sogan(Decorator):
    def __init__(self, component):
        self.component = component
        self.description = "Sogan"
        self.cost = malzeme_fiyatları["Sogan"]

class Misir(Decorator):
    def __init__(self, component):
        self.component = component
        self.description = "Misir"
        self.cost = malzeme_fiyatları["Misir"]


def display_menu():
    with open('Menu.txt', 'r') as f:
        print(f.read())


def order_start():
 
    display_menu()

   
    pizza = None
    while pizza is None:
        pizza_choice = input("Pizza secin: ")
        if pizza_choice == '1':
            pizza = KlasikPizza()
        elif pizza_choice == '2':
            pizza = MargaritaPizza()
        elif pizza_choice == '3':
            pizza = TurkPizza()
        elif pizza_choice == '4':
            pizza = DominosPizza()
        else:
            print("Secim hatali")    
    
   
    malzemeler = []
    malzeme_secim = None
    print("Malzeme secin : ")
    while malzeme_secim != '0':
      malzeme_secim = input("seciminizi girin: ")
      if malzeme_secim == '101':
        
        if not any(isinstance(s, Zeytin) for s in malzemeler):
            malzemeler.append(Zeytin(pizza))
        else:
            print("Zeytin zaten secili")
      elif malzeme_secim == '102':
        if not any(isinstance(s,Et) for s in malzemeler):
            malzemeler.append(Et(pizza))
        else:
            print("Et zaten secili")
      elif malzeme_secim == '103':
        if not any(isinstance(s, Kasar_peyniri) for s in malzemeler):
            malzemeler.append(Kasar_peyniri(pizza))
        else:
            print("Kasar peyniri zaten secili")
      elif malzeme_secim == '104':
        if not any(isinstance(s, Salam) for s in malzemeler):
            malzemeler.append(Salam(pizza))
        else:
            print("Salam zaten secili")
      elif malzeme_secim == '105':
        if not any(isinstance(s, Sogan) for s in malzemeler):
            malzemeler.append(Sogan(pizza))
        else:
            print("Sogan zaten secili")
      elif malzeme_secim == '106':
        if not any(isinstance(s, Mısır) for s in malzemeler):
            malzemeler.append(Mısır(pizza))
        else:
            print("Mısır zaten secili")
      elif malzeme_secim == '0':
        break
      else:
        print("Tekrar girin")

    
    pizza_fiyat = pizza.get_cost()
    PIZZA = pizza.get_description()
    description =','.join(s.get_description() for s in malzemeler)
    description = description.replace(PIZZA, "")
    malzeme_ucreti =  (sum(s.get_cost() for s in malzemeler) - (len(malzemeler)*pizza_fiyat))
    UCRET = malzeme_ucreti + pizza_fiyat
   
    if len(malzemeler) == 0:
      description = "None"
    
    print("\n","-"*15,"\n | malzeme ucreti: ",malzeme_ucreti,"₺ \n | Pizza fiyat: ",pizza_fiyat,"₺ \n","-"*15)
    print("\n"," *"*15,"\n Order: \n Pizza: ",PIZZA, "\n malzemeler:",description,"\n UCRET: ", UCRET, "₺","\n", " *"*15,"\n")

   
    confirmation = None
    while confirmation is None:
        confirmation_x = input("siparisi onaylayin (E or H): ")
        if confirmation_x == 'e' or confirmation_x == 'E':
          confirmation = "e"
          AD = input("Adınız: ")
          KREDIKARTINUMARASI = input("Kredi kartı numaranızı girin: ")
          PAROLA = input("Parolanızız girin: ")
          print("\n","* "*15,"\n",AD,"Odemeniz alındı. \n","* "*15)
          ZAMAN = datetime.datetime.now()
          ZAMAN.strftime("%Y-%m-%d %H:%M:%S")
          data=[[AD ,PIZZA, KREDIKARTINUMARASI, PAROLA, UCRET, ZAMAN]]

          
          with open('Orders_Database.csv', mode='a', newline='') as file:
            writer = csv.writer(file)
            writer.writerows(data)

        elif confirmation_x == 'h' or confirmation_x == 'H':
          confirmation = "h"
          print("Siparişiniz alınmadı")

        else:
          print("Şifre yanlış. Tekrar girin")

   
```

```python
order_start()
```
